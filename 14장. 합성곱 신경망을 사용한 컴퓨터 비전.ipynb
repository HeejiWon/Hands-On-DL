### 5. 케라스를 사용해 ResNet-34 CNN 구현하기
<img src='img/14_5.png' width='400'>
<img src='img/14_4.png' width='400'>

```python
# Residual Unit 층 만들기

class ResidualUnit(keras.layers.Layer):
    def __init__(self, filters, strides=1, activation='relu' , **kwargs):
        super().__init__(**kwargs)
        self.activation = keras.activations.get(activation)
        self.main_layers=[     # 그림 14-18의 오른쪽 모듈
            keras.layers.Conv2D(filters, 3, strides=strides,
                               padding='same', use_bias=False), # bias 안쓰는 이유는 아래에 배치 정규화층이 있기 때문
            keras.layers.BatchNormalization(),
            self.activation,
            keras.layers.Conv2D(filters, 3, strides=1,
                               padding='same', use_bias=False),
            keras.layers.BatchNormalization()]
        self.skip_layers = []  # 그림 14-18의 왼쪽 모듈
        
        if strides > 1:   # stride가 1보다 큰 경우, 결과값과 입력값을 연결할 때,
                          # 크기차이가 나므로 스트라이드가 같은 합성곱층 통과시키기
            self.skip_layers = [
                keras.layers.Conv2D(filters, 1, strides=strides,
                                   padding='same', use_bias=False),
                keras.layers.BatchNormalization()]
            
    def call(self, inputs):
        Z = inputs
        for layer in self.main_layers:
            Z = layer(Z)
        skip_Z = inputs
        for layer in self.skip_layers:
            skip_Z = layer(skip_Z)
        return self.activation(Z + skip_Z)
```

```python
model = keras.models.Sequential()
model.add(keras.layers.Conv2D(64, 7, strides=2, input_shape=[224, 224, 3],
                             padding='same', use_bias=False))
model.add(keras.layers.BatchNormalization())
model.add(keras.layers.Activation('relu'))
model.add(keras.layers.MaxPool2D(pool_size=3, strides=2, padding='same'))

prev_filters = 64
for filters in [64] * 3 + [128] * 4 + [256] * 6 + [512] * 3:
    strides = 1 if filters == prev_filters else 2
    model.add(ResidualUnit(filters, strides=strides))
    prev_filters = filters

model.add(keras.layers.GlobalAvgPool2D())
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(10, activation='softmax'))
```

### 7. 사전훈련된 모델을 사용한 전이 학습

```python
import tensorflow_datasets as tfds
```

```python
# 데이터 적재

dataset, info = tfds.load('tf_flowers', as_supervised=True, with_info=True)
                                                    # with_info : 데이터셋 정보
```

```python
info
```

```python
dataset_size = info.splits['train'].num_examples  # 3670
class_names = info.features['label'].names   # ['dandelion', 'daisy', ...]
n_classes = info.features['label'].num_classes  # 5
```

```python
# train test split 

test_set_raw, valid_set_raw, train_set_raw = tfds.load(
    "tf_flowers",
    split=["train[:10%]", "train[10%:25%]", "train[25%:]"],
    as_supervised=True)
```

```python
# 이미지 resize 하기

def preprocess(image, label):
    resized_image = tf.image.resize(image, [224, 224])
    final_image = keras.applications.xception.preprocess_input(resized_image) # 전처리
    return final_image, label
```

```python
# train set 섞은 후, 전처리
# 그다음 배치크기를 지정하고 프리패치를 적용

batch_size = 32
train_set = train_set_raw.shuffle(1000)
train_set = train_set.map(preprocess).batch(batch_size).prefetch(1)
valid_set = valid_set_raw.map(preprocess).batch(batch_size).prefetch(1)
test_set = test_set_raw.map(preprocess).batch(batch_size).prefetch(1)

# tf.image.random_crop() 함수로 이미지를 랜덤하게 자르거나
# tf.image.random_flip_left_right() 함수로 이미지를 수평으로 랜덤하게 뒤집을 수 있음
```

- keras.preprocessingimage.ImageDataGenerator 클래스를 사용하면 쉽게 디스크에서 이미지를 적재하여 여러 방식으로 데이터를 증식할 수 있음 ex) 이동, 회전, 크기 변환, 수평 뒤집기, 수직 뒤집기, 전단 변환
- tf.data 파이프라인으로 하는 것도 병렬로 처리하거나 요구에 맞게 조작할 수 있어 장점이 많음

```python
# Xception 모델 로드하고 모델 출력을 바탕으로 
# 새로운 전역 평균 풀링 층과 밀집 출력 층(+softmax) 만들기

base_model = keras.applications.xception.Xception(weights='imagenet',
                                                 include_top=False) 
    # include_top=False 지정하여 최상층 (전역 평균 풀링 층과 밀질 출력 층) 제외
avg = keras.layers.GlobalAveragePooling2D()(base_model.output)
output = keras.layers.Dense(n_classes, activation='softmax')(avg)
model = keras.Model(inputs=base_model.input, outputs=output)
```

```python
# 사전훈련된 층의 가중치 동결하기

for layer in base_model.layers:
    layer.trainable = False

# base_model 객체 자체를 사용하는 것이 아니라 기반 모델의 층을 직접 사용하기 때문에
# base_model.trainable=False로 지정해도 효과가 없음
```

```python
# 모델 컴파일하고 훈련 시작
# GPU 사용하지 않으면 매우 느림

optimizer = keras.optimizers.SGD(lr=0.2, momentum=0.9, decay=0.01)
                                # decay는 업데이트마다 적용되는 학습율 감소율
model.compile(loss='sparse_categorical_crossentropy', optimizer=optimizer,
             metrics=['accuracy'])
history = model.fit(train_set, epochs=5, validation_data=valid_set)

# 몇 번의 에포크 동안 75~80% 정확도, 더 나아지진 않음
# => 새로 추가한 최상위 층이 잘 훈련되었다는 걸 의미 (-> 동결 해제)
```

```python
# 사전훈련된 층 동결해제하고 훈련 계속하기
# 층 동결이나 해제할 때 모델 다시 컴파일 하기!!

for layer in base_model.layers:
    layer.trainable = True

optimizer = keras.optimizers.SGD(lr=0.01, momentum=0.9, decay=0.001)
model.compile(loss='sparse_categorical_crossentropy', optimizer=optimizer,
             metrics=['accuracy'])
history = model.fit(train_set, epochs=5, validation_data=valid_set)

# 정확도 95%
```
